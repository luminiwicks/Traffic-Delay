```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pd.read_csv(r'DataExtract.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#Splitting the data into independent and dependent variables
X = df.iloc[:,0:6].values
y = df.iloc[:,6:7].values
print('The independent features set: ')
print(X[:5,:])
print('The dependent variable: ')
print(y)
```

```python
#split the dataframe into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=9)
```

```python
lin_reg_mod = LinearRegression()
```

```python
#fit the model on the training data:
lin_reg_mod.fit(X_train, y_train)
print('Intercept: \n', lin_reg_mod.intercept_)
print('Coefficients: \n', lin_reg_mod.coef_)
```

```python
pred = lin_reg_mod.predict(X_test)
```

```python
pred
```

```python
#check the predictions against the actual values by using the RMSE and R-2 metrics, two metrics commonly used to evaluate regression tasks
test_set_rmse = (np.sqrt(mean_squared_error(y_test, pred)))

test_set_r2 = r2_score(y_test, pred)
```

```python
## Note that for rmse, the lower that value is, the better the fit
print(test_set_rmse)
# The closer towards 1, the better the fit
print(test_set_r2)
```
